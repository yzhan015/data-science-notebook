```python
df.drop_duplicates(inplace=True)
```

```python
import time
t = time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(3085992215))
t
```

```python
转换time这一列为年月日格式，并且作为一列加入表格中
```

```python
real_time = []
for i in df['time']:
    t = time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(i))
    real_time.append(t)
df['real_time'] = real_time
df
```

将real_time拆分成hour，date

```python
df['real_time'] = pd.to_datetime(df['real_time'] )
df['hour'] = df.real_time.dt.hour
df['date'] = df.real_time.dt.date
df['date'] = df['date'].astype('str')
df.head()
```

```python
df.date.min()
```

```python
删除2月份时间
```

```python
df.drop(labels=df.loc[df.date == '2067-02-21'].index,inplace=True)
```

```python
df.reset_index(drop=True,inplace=True)
```

计算抖音播放量来源分布

```python
channel = df.groupby(by='channel').count()['uid']
channel
```

作品时长与播放量的关系

```python
item_duration = df.groupby(by = 'duration_time').count()['uid']
item_duration
```

```python
plt.plot(item_duration)
plt.xlim(2,40)
plt.xlabel('作品时长')
plt.ylabel('播放量')
plt.title('作品时长和播放量的关系图')
```

计算作品时长与作品数量的关系

```python
duration_nums = df.groupby(by='duration_time').count()['item_id'].unique()
duration_nums
```

```python
plt.plot(duration_nums)
plt.xlim(2,40)
plt.xlabel('作品时长')
plt.ylabel('作品数量')
plt.title('作品时长与作品数量的关系图')
```

计算作品时长与完播率的关系

```python
duration_finish = df.groupby(by='duration_time').count()['finish'].mean()
duration_finish
```

```python
plt.plot(duration_finish)
plt.xlim(2,40)
plt.xlabel('作品时长')
plt.ylabel('完播率')
plt.title('作品时长与完播率的关系')
```

```python
like_p = df.groupby('duration_time')['like'].mean()
plt.plot(like_p)
plt.xlim(2,40)
plt.xlabel('作品时长')
plt.ylabel('点赞率')
plt.title("作品时长与点赞率的关系")
```

作品发布时间（小时）与点赞完播率之间的关系

```python
h = df.groupby(by='hour')[['finish','like']].mean()
h
```

```python
plt.plot(h)
plt.title('作品发布时间与点赞完播率之间的关系')
```

发布作品时间和播放量之间的关系

查找被作品使用次数最多的前10首音乐id

```python
music_ = df.groupby(by='music_id')['uid'].count().sort_values(ascending=False)
top_10_music = music_.iloc[:10]
top_10_music
```

```python
plt.plot(top_10_music.index.astype('str'),top_10_music.values)
plt.title('')
```
